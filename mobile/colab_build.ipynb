```python
# 5. 安装 akshare 和 beautifulsoup4 (在 buildozer 环境中)
!pip install akshare beautifulsoup4

# 开始打包 APK（此步骤约需 20-30 分钟）
# 清理之前的构建缓存，确保使用新的依赖
!rm -rf .buildozer
!buildozer -v android debug
```

```python
# 6. 显示打包结果
from google.colab import files
import glob

apk_files = glob.glob('bin/*.apk')
if apk_files:
    print(f"找到 APK 文件：{apk_files[0]}")
    files.download(apk_files[0])
    print("下载已开始！")
else:
    print("打包失败，请检查上方日志")
```
